## Data Preprocessing

```python
def data_preprocess (df : pd.DataFrame) -> pd.DataFrame :
    
    df['Age'] = df['Age'].fillna(df['Age'].mean())
    df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
    df['Embarked'] = df['Embarked'].fillna('S')
    
    df = df.drop(columns = ['Name'])
    
    return df
```

```python
def cabin_extraction (df : pd.DataFrame) -> pd.DataFrame :
    
    # Sample data
    data = df['Cabin']

    # Split the data into two new Series
    letters = data.str.extract(r'([A-Za-z]+)', expand=False)
    numbers = data.str.extract(r'(\d+)', expand=False).astype(float)

    # Fill NaNs in the letters series with the mode
    mode_letter = 'X'  # mode() returns a Series, we take the first value
    letters = letters.fillna(mode_letter)

    # Fill NaNs in the numbers series with 0
    numbers = numbers.fillna(0).astype(int)

    df['Cabin'] = letters
    df['Cabin_no'] = numbers
    
    return df
```

```python
train_df = cabin_extraction(data_preprocess(train_df))
test_df = cabin_extraction(data_preprocess(test_df))
```

```python
show_columns_counts(train_df)
show_columns_counts(test_df)
```
